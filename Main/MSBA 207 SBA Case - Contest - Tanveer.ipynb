## “Should This Loan be Approved or Denied?”<br>Predictive Modeling Using the SBA National Data<br>
The U.S. Small Business Administration (SBA) was founded in 1953 on the principle of promoting<br>
and assisting small enterprises in the U.S. credit market (SBA Overview and History, US Small<br>
Business Administration (2015). Small businesses have been a primary source of job creation in<br>
the United States; therefore, fostering small business formation and growth has social benefits by<br>
creating job opportunities and reducing unemployment. One way SBA assists these small business<br>
enterprises is through a loan guarantee program which is designed to encourage banks to grant<br>
loans to small businesses. SBA acts much like an insurance provider to reduce the risk for a bank<br>
by taking on some of the risk through guaranteeing a portion of the loan. In the case that a loan<br>
goes into default, SBA then covers the amount they guaranteed.<br><br>
There have been many success stories of start-ups receiving SBA loan guarantees such as FedEx<br>
and Apple Computer. However, there have also been stories of small businesses and/or start-ups<br>
that have defaulted on their SBA-guaranteed loans. The rate of default on these loans has been a<br>
source of controversy for decades. Conservative economists believe that credit markets perform<br>
efficiently without government participation. Supporters of SBA-guaranteed loans argue that the<br>
social benefits of job creation by those small businesses receiving government-guaranteed loans<br>
far outweigh the costs incurred from defaulted loans.<br><br>
Since SBA loans only guarantee a portion of the entire loan balance, banks will incur some losses<br>
if a small business defaults on its SBA-guaranteed loan. Therefore, banks are still faced with a<br>
difficult choice as to whether they should grant such a loan because of the high risk of default. One<br>
way to inform their decision-making is through analyzing relevant historical data such as the SBA<br>
National Data with the following Data Dictionary (see Li, Mickel, and Taylor (2018) for<br>
background information about the data and analysis):<br><br>
Data Dictionary<br>
NAME: National SBA<br>
TYPE: Census<br>
SIZE: `899,164` observations, `27` variables<br>
SOURCE: United States Small Business Administration<br><br>
STORY BEHIND THE DATA: This data set is from the U.S. Small Business Administration<br>
(SBA) and provides historical data from 1987 through 2014, containing 27 variables and 899,164<br>
observations. Each observation represents a loan that was guaranteed to some degree by the SBA.<br>
Included is a variable [MIS_Status] which indicates if the loan was paid in full or<br>
defaulted/charged off.<br>

### Discriminant Analysis. (python 3.10.11)

Utilizing the preprocessing and analysis above we are creating a subset_df for the optimised predictors.

```python
predictors_da= ['State','NAICS',
 'Term',
 'CreateJob',
 'RetainedJob',
 'UrbanRural',
 'RevLineCr',
 'DisbursementGross']
df_SBA_DA = df_SBA_mod_1[outcome+predictors_da]
```

```python
X = pd.get_dummies(df_SBA_DA[predictors_da], drop_first=True)
y = df_SBA_DA['MIS_Status']
X_train,X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.6, random_state=1)
print(X_train.shape)
print(X_valid.shape)
```

```python
sba_lin = LinearDiscriminantAnalysis()
sba_lin.fit(X_train, y_train)
```

With the info that The costs of incorrectly classifying a loan application as lower risk outweigh the benefits of correctly classifying a loan application as lower risk by a factor of 5 we will adjust the priors to refelct this and rerun the model before checking its performance. 

```python
sba_lin.priors_
```

```python
low_risk_prior = sba_lin.priors_[0]
high_risk_prior = sba_lin.priors_[1]
low_risk_prior_new = (low_risk_prior)/((low_risk_prior)+ (high_risk_prior * 5))
high_risk_prior_new = (high_risk_prior * 5)/((low_risk_prior)+ (high_risk_prior * 5))
```

```python
sba_lin = LinearDiscriminantAnalysis(priors=[low_risk_prior_new, high_risk_prior_new])
sba_lin.fit(X_train, y_train)
classificationSummary(y_valid, 
                      sba_lin.predict(X_valid),
                      class_names=['NonDefault', 'Default'])
```

```python
proba = sba_lin.predict_proba(X_valid)
result = pd.DataFrame({'actual': y_valid, 
                       'p(0)': [p[0] for p in proba],
                       'p(1)': [p[1] for p in proba],
                       'predicted': sba_lin.predict(X_valid) })
df = result.sort_values(by=['p(1)'], ascending=False)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
df['actual'] = df['actual'].astype('int64')
gainsChart(df.actual, ax=axes[0])
liftChart(df.actual, title=False)
plt.show()
```
